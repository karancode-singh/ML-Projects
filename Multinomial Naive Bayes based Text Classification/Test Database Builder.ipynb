```python
#Routine stuff
import numpy as np
import pandas as pd
from string import punctuation
import os
import operator
import matplotlib.pyplot as plt
```

```python
#import vocab
data = pd.read_csv("vocab.csv",delimiter=",")
data = data.drop('Unnamed: 0',axis=1)
```

```python
K = 2000 #limit of vocab considered
vocab = list(data[:K]['0'])
vocab
```

```python
columns = vocab
columns.append('CLASSIFICATION')
columns
```

```python
df = pd.DataFrame(columns=columns) #make empty dataframe with headings as vocab words & classification
df
```

```python
#iterating just like done in vocabulary builder
#refer there for the comments
tempPath = "/Users/jo/Documents/CN_ML/Text Classification - Project/20_newsgroups"
os.chdir(tempPath)
countDir = 0
totalDir = len(os.listdir(os.getcwd()))
for directory in os.listdir(os.getcwd()):
    if directory[0]=='.':
        continue
    path = tempPath + "/" + directory
    os.chdir(path)
    files = os.listdir(os.getcwd())
    for file in files:
        if file[0]=='.':
            continue
    n = len(files)
    n_train = int(n*0.8) #number of training data
    #n_test = n-n_train
    #files = files[:n_train]
    files = files[n_train:] #selecting only testing files (opposite of training files)
    count = 0
    print(n-n_train, "files to process in", directory)
    dirData = [[0]*len(columns)]
    for file in files:
        row_data = [0]*len(columns)
        f = open(file,"r",errors="ignore")
        for word in f.read().split():
            word = word.strip(punctuation).lower()
            #if word not in stop_words and word not in ignore_words and word.isalpha() and len(word)!=1:
            if word in vocab:
                row_data[columns.index(word)] = row_data[columns.index(word)] + 1
        row_data[columns.index('CLASSIFICATION')] = directory
        dirData.append(row_data)
        count += 1
        if count%100 == 0:
            print("Done with files: ", count)
    tempDF = pd.DataFrame(dirData[1:], columns=columns)
    df = df.append(tempDF, ignore_index=True)
    countDir += 1
    print("Done with directory: ", directory)
    print("Number: ", countDir, "of", totalDir)
    print()
```
